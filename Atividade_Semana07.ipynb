###Questão 1

Importe a planilha `dados_starwars.xlsx` na pasta virtual `content`.

```python
# Questao 1
import pandas as pd
dados_starwars = pd.read_excel("dados_starwars.xlsx")
dados_starwars
```

###Questão 2

Remova todos os valores faltantes. Salve a base de dados atualizada. 

```python
# Questao 2
#verificando quantidade de linhas e colunas antes de dropar os valores vazios
print(dados_starwars.shape)

#dropando/removendo valores vazios
dados_starwars = dados_starwars.dropna()

#verificando quantidade de linhas e colunas depois de dropar os valores vazios
print(dados_starwars.shape)
```

###Questão 3

Filtre somente os registros com `especie` = `humano`. Salve a base de dados atualizada. 

```python
# Questao 3
dados_starwars = dados_starwars[dados_starwars["especie"] == "Humano"]
dados_starwars
```

###Questão 4

Edite a variável `altura` (que esta em centímetros) para que passe para a unidade em metros (*dividir por 100*). Salve a base de dados atualizada. 

```python
# Questao 4

dados_starwars["altura"] = dados_starwars["altura"]/100
dados_starwars

```

###Questão 5

Renomeie a variável `altura` para `altura_metros`. Salve a base de dados atualizada. 

```python
# Questao 5
dados_starwars.rename(columns={'altura': 'altura_metros'}, inplace = True)
dados_starwars
```

###Questão 6

Ordene a base de dados por `altura_metros`. Salve a base de dados atualizada. 

```python
# Questao 6
dados_starwars = dados_starwars.sort_values(by='altura_metros')
dados_starwars
```
