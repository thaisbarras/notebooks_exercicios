```python
# Questao 1

# Carregando pacotes
import pandas as pd
import ipeadatapy

# Obtendo serie com dados sobre emplacamento de veículos
emplacamentos = ipeadatapy.timeseries('FENABRAVE12_VENDAUTO12')

# A função Python Pandas DataFrame.reset_index() redefine o índice do quadro de dados fornecido. 
#Ele substitui o índice antigo pelo índice padrão. Se o quadro de dados fornecido tiver um MultiIndex, esse método removerá todos os níveis.
emplacamentos.reset_index(inplace=True)

emplacamentos = emplacamentos.loc[(emplacamentos["DATE"] >= "2020-01-01") & (emplacamentos["DATE"] < "2021-01-01"), "VALUE (Unidade)"]
emplacamentos.rename("emplacamentos", inplace=True)
emplacamentos
```

```python
# ---------- TRECHO DE TESTE ---------
# Para aprimorar meus conhecimentos adicionei essa célula para fazer testes, 
# mas não há necessidade de executá-la.

# Não há necessidade de ordenar os valores, por os métodos estatísticos do Python fazem os cálculos considerando a ordenação.
emplacamentos = emplacamentos.sort_values()
emplacamentos
```

```python
# Questao 1 (continuacao)

# Calcule as estatísticas de interesse
# média, mediana, mínimo, máximo, amplitude, variância e desvio padrão.



media = round(emplacamentos.mean(), 2)
mediana = emplacamentos.median()
minimo = emplacamentos.min()
maximo = emplacamentos.max()
amplitude = maximo - minimo
variancia = round(emplacamentos.var(), 2)
desvio_padrao = round(emplacamentos.std(), 2)

print(f"média: {media}")
print(f"mediana: {mediana}")
print(f"mínimo: {minimo}")
print(f"máximo: {maximo}")
print(f"amplitude: {amplitude}")
print(f"variância: {variancia}")
print(f"desvio padrão: {desvio_padrao}")


# Como dito na aula, verifiquei os cálculos multiplicando o desvio padrão por 2 e somando à média.
# O valor encontrado foi 233637,33, isso significa que os cálculos da mádia e desvio-padrão estão coerentes
```

###Questão 3

O Imposto sobre Produtos Industrializados, cuja sigla é IPI, é um imposto federal, ou seja, somente a União pode instituí-lo ou modificá-lo, sobre produtos industrializados no Brasil. Abaixo, obtemos os dados da arrecadação do IPI sobre automóveis em 2022.

Calcule a correlação entre o número de veículos emplacados e o IPI arrecadado em 2022.

```python
# Questao 3

# Obtendo serie com dados sobre IPI arrecadado sobre automóveis
ipi = ipeadatapy.timeseries('SRF12_IPIVEI12')
ipi.reset_index(inplace=True)
ipi = ipi.loc[(ipi["DATE"] >= "2020-01-01") & (ipi["DATE"] < "2021-01-01"), 
                                  "VALUE (R$)"]
ipi.rename("ipi", inplace=True)

carros = pd.concat([emplacamentos.reset_index(drop=True),
                    ipi.reset_index(drop=True)],
                   axis=1)
carros
```

```python
# Questao 3 (continuacao)

# Calcule a correlacao entre o numero de emplacamentos e o IPI recolhido pela União.
correlacao = carros.corr()
correlacao
```

###Questão 4

Com a base de dados `carros`, faça um gráfico de dispersão utilizando `plt.scatter` entre `emplacamentos` e `ipi`.

```python
# Questao 4
from matplotlib import pyplot as plt

# Grafico de dispersao

plt.scatter(carros['emplacamentos'], carros['ipi'], c='k', label="Emplacamentos x IPI")


plt.legend(loc="best")
plt.xlabel("Número de emplacamentos")
plt.ylabel("VIPI")
plt.title("Correlacao entre o Número de emplacamentos e o IPI recolhido pela União.")


# Exibe o gráfico
plt.show()
```

```python
# testando Numpy que aprendi um outro curso
import numpy as np

#Criando funções de 1°, 2°, e 3° grau

# A classe polyfit usa o método dos mínimos quadrados para ajustar um polinômio de grau “deg” (1, 2, 3 ...) em um conjunto de dados emplacamentos vs ipi. 
primeiro_grau = np.polyfit(emplacamentos, ipi, 1)
segundo_grau = np.polyfit(emplacamentos, ipi, 2)
terceiro_grau = np.polyfit(emplacamentos, ipi, 3)

# Já a classe poly1d permite calcular o novo vetor ipi. (valores ajustados para cada valor de emplacamentos).
# A função numpy.poly1d() ajuda a definir uma função polinomial. Facilita a aplicação de “operações naturais” em polinômios.
p1 = np.poly1d(primeiro_grau)
p2 = np.poly1d(segundo_grau)
p3 = np.poly1d(terceiro_grau)

plt.scatter(carros['emplacamentos'], carros['ipi'], c='k', label="Emplacamentos x IPI")

#numpy
x = np.linspace(emplacamentos.min(), ipi.max(), 200)
plt.plot(x, p1(x), c='r', label="Regressão Grau 1")
plt.plot(x, p2(x), color='blue', label="Regressão Grau 2")
plt.plot(x, p3(x), c='g', label="Regressão Grau 3")

plt.legend(loc="best")
plt.xlabel("Número de emplacamentos")
plt.ylabel("VIPI")
plt.title("Correlacao entre o Número de emplacamentos e o IPI recolhido pela União.")


# Exibe o gráfico
plt.show()
```
